```python
# case 9, x direction
to = 0
vx = 0
xo = 0
tin = [
    2, 6, 2,
    3, 4, 3,
    2, 6, 2, 
    2, 1, 2,
    1, 0, 1,
    0]
ax = [
    2, 0, -2,
    0, 0, 0,
    -2, 0, 2,
    0, 0, 0,
    10, 0, -10,
    0]
motion_x = PiecewiseMotion(to, vx, xo, tin, ax)
motion_x.plot(resolution=1000)
```

```python
# case 9, y direction
to = 0
vy = 0
yo = 0
tin = [
    2, 6, 2,
    3, 4, 3,
    2, 6, 2,
    2, 1, 2,
    1, 0, 1,
    0]
ay = [
    0, 0, 0,
    2, 0, -2,
    0, 0, 0,
    -2, 0, 2,
    -10, 0, 10,
    0]
motion_y = PiecewiseMotion(to, vy, yo, tin, ay)
motion_y.plot(resolution=1000)
```

```python
import numpy as np

to = 0
tin = [
    2, 6, 2,
    3, 4, 3,
    2, 6, 2,
    2, 1, 2,
    1, 0, 1,
    0]
res = 1000
tt_vals = np.linspace(to, to + sum(tin), res)

ax_vals = [motion_x.get_acceleration(t) for t in tt_vals]
vx_vals = [motion_x.get_velocity(t) for t in tt_vals]
rx_vals = [motion_x.get_position(t) for t in tt_vals]

ay_vals = [motion_y.get_acceleration(t) for t in tt_vals]
vy_vals = [motion_y.get_velocity(t) for t in tt_vals]
ry_vals = [motion_y.get_position(t) for t in tt_vals]

```

```python
import matplotlib.pyplot as plt

plt.subplot(2, 2, 1)
plt.plot(tt_vals, ry_vals, 'b')
plt.xlabel('t')
plt.ylabel('y')
plt.grid()

plt.subplot(2, 2, 4)
plt.plot(rx_vals, tt_vals, 'g')
plt.xlabel('x')
plt.ylabel('t')
plt.grid()

plt.subplot(2, 2, 2)
plt.plot(rx_vals, ry_vals, 'r')
plt.grid()
plt.title('y -- x')

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.subplot(2, 2, 1)
plt.plot(tt_vals, vy_vals, 'b')
plt.xlabel('t')
plt.ylabel('vy')
plt.grid()

plt.subplot(2, 2, 4)
plt.plot(vx_vals, tt_vals, 'g')
plt.xlabel('vx')
plt.ylabel('t')
plt.grid()

plt.subplot(2, 2, 2)
plt.plot(vx_vals, vy_vals, 'r')
plt.grid()
plt.title('vy -- vx')

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.subplot(2, 2, 1)
plt.plot(tt_vals, ay_vals, 'b')
plt.xlabel('t')
plt.ylabel('ay')
plt.grid()

plt.subplot(2, 2, 4)
plt.plot(ax_vals, tt_vals, 'g')
plt.xlabel('ax')
plt.ylabel('t')
plt.grid()

plt.subplot(2, 2, 2)
plt.plot(ax_vals, ay_vals, 'r')
plt.grid()
plt.title('ay -- ax')

plt.show()
```
